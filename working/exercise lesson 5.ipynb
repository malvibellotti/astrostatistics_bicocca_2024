```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
n_draws = 1000
x_i = np.random.normal(loc=1, scale=0.2, size=n_draws)
x = np.linspace(0, 2, 100)
```

```python
product = np.ones(len(x))
sum_log = np.zeros(len(x)) 
for mu in x_i:
    p_xi = norm(mu, 0.2)
    plt.plot(x, p_xi.pdf(x), c='green')
    product = p_xi.pdf(x) * product
    sum_log = sum_log + np.log(p_xi.pdf(x))
    
```

```python
plt.plot(x, product, c='red', ) #label=r'$\mu=5,\ \sigma=1$'
index_max= np.argmax(product)
x_max = x[index_max]
print(x_max)
plt.axvline(x_max,c='red',ls='dotted')
```

```python
likelihood = np.exp(sum_log)
plt.plot(x, sum_log, c='red', ) #label=r'$\mu=5,\ \sigma=1$'
index_max_sum = np.argmax(likelihood)
x_max_sum= x[index_max_sum]
print(x_max)
plt.axvline(x_max_sum,c='red',ls='dotted')
```
